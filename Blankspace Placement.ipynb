## DATA VISUALIZATION 

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = product1['age']
y = product1['purchase_count']

w = product2['age']
z = product2['purchase_count']


plt.figure(figsize=(12,8))
plt.scatter(x,y, label = 'Product 1')
plt.scatter(w,z, alpha = 0.8, label = 'Product 1')
plt.xlabel('Age')
plt.ylabel('Total Purchased')
plt.title('Total Purchased Product Based On Age')
plt.legend()
plt.show()
```

```python
total = product1['purchase_count'].sum() + product2['purchase_count'].sum() 
size1 = int(product1['purchase_count'].sum() / total * 100)
size2 = int(product2['purchase_count'].sum() / total * 100)

size = [size1,size2]
label = 'Product 1' , 'Product 2'
explode = (0, 0.1)


plt.figure(figsize=(12,8))
plt.pie(size, labels = label, autopct='%1.2f%%', explode =explode)
plt.title('Ratio Sales Each Product')
plt.show()
```

```python
labels = ['Product']
product_1_revenue = product1['total_paid'].sum()
product_2_revenue = product2['total_paid'].sum()

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize = (12,8))
rects1 = ax.bar(x - width/2, product_1_revenue, width, label='Product 1')
rects2 = ax.bar(x + width/2, product_2_revenue, width, label='Product 2')
                     
ax.set_ylabel('Revenue (IDR)')
ax.set_title('Total Revenue Each Product')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_yticks([10000000000,20000000000,30000000000,40000000000,50000000000])
ax.set_yticklabels(['10 B','20 B','30 B','40 B','50 B'])
ax.legend()


fig.tight_layout()

plt.show()
```

```python
channel1 = product1.groupby('channel')['purchase_count'].sum().reset_index().sort_values('channel', ascending=False)
channel2 = product2.groupby('channel')['purchase_count'].sum().reset_index().sort_values('channel', ascending=False)

labels = ['ONLINE','OFFLINE']
product_1_channel = channel1['purchase_count']
product_2_channel = channel2['purchase_count']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize = (12,8))
rects3 = ax.bar(x - width/2, product_1_channel, width, label='Product 1')
rects4 = ax.bar(x + width/2, product_2_channel, width, label='Product 2')
                     
ax.set_ylabel('Times (x)')
ax.set_title('Total Purchased Product Based On Channel')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()

plt.show()
```
